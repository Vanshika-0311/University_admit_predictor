```python
x= df.iloc[:,0:7]
x
```

```python
y= df.iloc[:,7:]
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=9)
```

```python
from sklearn.linear_model import LinearRegression
MLR = LinearRegression()
MLR.fit(x_train,y_train)
```

```python
y_pred=MLR.predict(x_test)
y_pred.reshape(1,-1)
```

```python
#R2-Score
from sklearn.metrics import accuracy_score,recall_score,roc_auc_score,confusion_matrix, r2_score, classification_report

r2_score_mlr = r2_score(y_test,y_pred)

print("Linear Regression's Score = {:.3f}".format(r2_score_mlr))
```

```python
#MSE
from sklearn import metrics
print(metrics.mean_squared_error(y_test,y_pred))
```

```python
#RMSE
print(np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
